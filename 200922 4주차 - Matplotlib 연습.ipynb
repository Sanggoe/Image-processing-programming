```python
from matplotlib import pyplot as plt
import numpy as np
a = np.array([2,5,4,3,12,10])
b = np.array([10,2,3,4,26,11])
plt.plot(a) # x축에는 인덱스 번호, y축에는 값을 그래프로 표시한다!
plt.plot(b)
plt.show() # 그래프 보여줘
```

```python
x = np.arange(10)
y = x**2
plt.plot(x,y,'g') # 그래프의 색깔 바꿀 수 있다! 보통 영어 단어 첫글자
plt.show()
```

```python
x=np.arange(10)
f1 = x * 5
f2 = x ** 2
f3 = x ** 2 + x * 2
plt.plot(x, 'r--') # 점선
plt.plot(f1, 'g.') # 점 점 점 점
plt.plot(f2, 'bv') # 역 삼각형
plt.plot(f3, 'ks') # 네모
plt.show()
```

```python
x = np.arange(10)

plt.subplot(2,2,1) # 바둑판 모양으로, 네 개 중에 첫 번째에다 써줘라
plt.plot(x, x**2, 'g')
plt.subplot(2,2,2)
plt.plot(x, x*5)
plt.subplot(223) # , 안써도 되지만, 내가 헷갈리자나~
plt.plot(x, np.sin(x))
plt.subplot(224)
plt.plot(x, np.cos(x))
plt.show()
```

```python
import cv2

img = cv2.imread('img/jiheon.jpg') # OpenCV는 기본적으로 Color로 인식!!
img2 = cv2.imread('img/jiheon.jpg') # 그레이 스케일을 지정할 수도 있다!!
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()
# 사진은 RGB가 아니라 BGR 순으로 저장된다. 따라서.. R과B가 반전되어 보이는 것이다..
# 그 값의 위치를 반전시켜주는 방법..!!
```

```python
plt.subplot(1,2,1)
plt.imshow(img[:,::-1,::-1]) # 모든 픽셀, 모든 칼라를 거꾸로 출력하라!
# x, y, color!
plt.xticks([]) # x축 비워버려라
plt.yticks([]) # y축 비워버려라

plt.subplot(1,2,2)
plt.imshow(img2[::-1,:,::-1]) # 모든 픽셀, 모든 칼라를 거꾸로 출력하라!
img2.fill(0)
plt.xticks([]) # x축 비워버려라
plt.yticks([]) # y축 비워버려라
plt.show()
```
